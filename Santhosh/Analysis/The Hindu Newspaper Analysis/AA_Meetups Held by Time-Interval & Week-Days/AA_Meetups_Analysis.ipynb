## Data Visualization on AA_Meetups held in the region and week during time interval of (2006 to 2019)

### Importing necessary Libraries

```python
import matplotlib.pyplot as plt # plotting library for the Python
import pandas as pd # primary data structure library
import numpy as np # general-purpose array-processin
import os, sys #  System-specific parameters and functions.
%matplotlib inline
```

### Collected data ['AA_Meets.csv'] read into a pandas dataframe

```python
AA_Meets = pd.read_csv('AA_Meets.csv')
AA_Meets.head()
```

### Creating DataFrame interms of time interval (2006 to 2019)

```python
y_2019 = AA_Meets[AA_Meets['Year']==2019]
y_2018 = AA_Meets[AA_Meets['Year']==2018]
y_2017 = AA_Meets[AA_Meets['Year']==2017]
y_2016 = AA_Meets[AA_Meets['Year']==2016]
y_2015 = AA_Meets[AA_Meets['Year']==2015]
y_2014 = AA_Meets[AA_Meets['Year']==2014]
y_2013 = AA_Meets[AA_Meets['Year']==2013]
y_2012 = AA_Meets[AA_Meets['Year']==2012]
y_2011 = AA_Meets[AA_Meets['Year']==2011]
y_2010 = AA_Meets[AA_Meets['Year']==2010]
y_2009 = AA_Meets[AA_Meets['Year']==2009]
y_2008 = AA_Meets[AA_Meets['Year']==2008]
y_2007 = AA_Meets[AA_Meets['Year']==2007]
y_2006 = AA_Meets[AA_Meets['Year']==2006]
```

### Creating the DataFrame of days with the time_interval (2006 to 2019)

```python
week_df = pd.DataFrame({'2019': pd.value_counts(y_2019['Day']),
                   '2018': pd.value_counts(y_2018['Day']),
                   '2017': pd.value_counts(y_2017['Day']),
                   '2016': pd.value_counts(y_2016['Day']),
                   '2015': pd.value_counts(y_2015['Day']),
                   '2014': pd.value_counts(y_2014['Day']),
                   '2013': pd.value_counts(y_2013['Day']), 
                   '2012': pd.value_counts(y_2012['Day']),
                   '2011': pd.value_counts(y_2011['Day']),
                   '2010': pd.value_counts(y_2010['Day']), 
                   '2009': pd.value_counts(y_2009['Day']), 
                   '2008': pd.value_counts(y_2008['Day']),
                   '2007': pd.value_counts(y_2007['Day']),
                   '2006': pd.value_counts(y_2006['Day']),})

week_df = week_df.fillna(0).astype(int)
```

### During the time interval (2006 to 2019) the total no.of AA Meetings held in the following Days
- Monday
- Tuesday
- Wednesday
- Thrusday
- Friday
- Saturday
- Sunday

```python
%%time
# AA Meetings in Chennai during time interval (2006-2019)
weedays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for index, item in enumerate(weedays):
    x = week_df.loc[str(item)].index.tolist()
    y = week_df.loc[str(item)].tolist()

    ffig, ax = plt.subplots()    
    width = 0.4 # the width of the bars 
    ind = np.arange(len(y))  # the x years for the groups
    ax.barh(ind, y, width)
    ax.set_yticks(ind+width/2)
    ax.set_yticklabels(x, minor=False)
    for i, v in enumerate(y):
        ax.text(v-0.5, i +0.1, " "+str(v), size = 10)
    plt.title('AA Meetings held on '+ str(item) +' during the time period 2006 to 2019')
    plt.xlabel('Total number of Meetings held on ' + str(item))
    plt.ylabel('Time Interval 2006 to 2019')  
    plt.savefig(os.path.join("D:\\!Code!\\CDRI\\#News_Testing\\The Hindu\\Week_Visuals\\"+str(item)+".png"), dpi=1000, format="png", bbox_inches="tight") # use format='svg' or 'pdf' for vectorial pictures
    
    sys.stdout.write('\r' + str(index) + ' : ' + str(item) + '\r')
    sys.stdout.flush()
```

### Creating the DataFrame of Published Region with the time_interval (2006 to 2019)

```python
region_df = pd.DataFrame({'2019': pd.value_counts(y_2019['Published_Region']),
                   '2018': pd.value_counts(y_2018['Published_Region']),
                   '2017': pd.value_counts(y_2017['Published_Region']),
                   '2016': pd.value_counts(y_2016['Published_Region']),
                   '2015': pd.value_counts(y_2015['Published_Region']),
                   '2014': pd.value_counts(y_2014['Published_Region']),
                   '2013': pd.value_counts(y_2013['Published_Region']), 
                   '2012': pd.value_counts(y_2012['Published_Region']),
                   '2011': pd.value_counts(y_2011['Published_Region']),
                   '2010': pd.value_counts(y_2010['Published_Region']), 
                   '2009': pd.value_counts(y_2009['Published_Region']), 
                   '2008': pd.value_counts(y_2008['Published_Region']),
                   '2007': pd.value_counts(y_2007['Published_Region']),
                   '2006': pd.value_counts(y_2006['Published_Region']),})

region_df = region_df.fillna(0).astype(int)
```

### During the time interval (2006 to 2019) the total no.of AA Meetings held in the following Published Region
- Chennai
- Madurai
- Coimbatore
- Thiruvananthapuram
- Tirunelveli
- Engagements

```python
%%time
location = ['chennai', 'coimbatore', 'engagements', 'madurai', 'thiruvananthapuram', 'tirunelveli']
for index, item in enumerate(location):
    x = region_df.loc[str(item)].index.tolist()
    y = region_df.loc[str(item)].tolist()

    ffig, ax = plt.subplots()    
    width = 0.4 # the width of the bars 
    ind = np.arange(len(y))  # the x years for the groups
    ax.barh(ind, y, width)
    ax.set_yticks(ind+width/2)
    ax.set_yticklabels(x, minor=False)
    for i, v in enumerate(y):
        ax.text(v-0.5, i +0.1, " "+str(v), size = 10)
    plt.title('AA Meetings held in '+ str(item) +' during the time interval 2006 to 2019')
    plt.xlabel('Total number of Meetings held in ' + str(item))
    plt.ylabel('Time Interval 2006 to 2019')  
    plt.savefig(os.path.join("D:\\!Code!\\CDRI\\#News_Testing\\The Hindu\\Published_Region_Visuals\\"+str(item)+".png"), dpi=1000, format="png", bbox_inches="tight") # use format='svg' or 'pdf' for vectorial pictures
    sys.stdout.write('\r' + str(index) + ' : ' + str(item) + '\r')
    sys.stdout.flush()
```
