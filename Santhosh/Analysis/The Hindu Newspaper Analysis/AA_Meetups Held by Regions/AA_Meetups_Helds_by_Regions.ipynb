```python
from plotnine import *
import pandas as pd
```

```python
df = pd.read_csv('AA_Meets.csv')
df.head()
```

```python
sources = df.groupby('Published_Region').agg('count')
source_counts = sources.Meetings.sort_values(ascending=False)
source_data = pd.DataFrame(source_counts).reset_index()
```

```python
sources_plot = ggplot(source_data, aes(x='Published_Region', y= 'Meetings', fill='Published_Region')) + geom_col() + coord_flip()\
+ scale_fill_brewer(type='div', palette="Spectral") + theme_classic() + ggtitle('AA_Meetups Schedule held in different regions')\
+ geom_text(aes(label='Meetings'), size=8, ha='right', format_string='{}')\

display(sources_plot)
```

```python
ggsave(plot=sources_plot, filename='region.png', height=5, width=5, units = 'in', dpi=1000)
```
