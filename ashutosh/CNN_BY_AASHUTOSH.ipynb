```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
```

```python

chromeOptions = webdriver.ChromeOptions()
chromeOptions.add_argument('--headless')
browser=webdriver.Chrome(options = chromeOptions, executable_path = r'/home/aashutosh/Documents/webScrapingStuffs/chr webdriver/chromedriver')

headlines_l = []
dates_l = []
body_l = []

for i in range(1,6):
    url = 'https://edition.cnn.com/search/?q=Alcoholics%20Anonymous&size=10&page=' + str(i)
    print(url)
    browser.get(url)
    time.sleep(1)
    headlines = browser.find_elements_by_class_name('cnn-search__result-headline')
    dates = browser.find_elements_by_class_name('cnn-search__result-publish-date')
    body = browser.find_elements_by_class_name('cnn-search__result-body')



    for head in headlines:
        print(head.text)
        headlines_l.append(head.text)
    for date in dates:
        print(head.text)
        dates_l.append(date.text)
    for bd in body:
        print(bd.text)
        body_l.append( bd.text)

data = pd.DataFrame([headlines_l,dates_l,body_l])
data.T
data.to_csv('cnn.csv',mode='w',header = False)
```
